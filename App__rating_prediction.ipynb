```python
#Preprocessing
from sklearn import metrics
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df['Price'] = df['Price'].apply(lambda x : x.strip('$'))
```

```python
df['Installs'] = df['Installs'].apply(lambda x : x.strip('+').replace(',', ''))
```

```python
category_list = df['Category'].unique().tolist()
category_list = ['cat_' + word for word in category_list]
df = pd.concat([df, pd.get_dummies(df['Category'], prefix='cat')], axis=1)
```

```python
df['Android Ver'] = df['Android Ver'].apply(lambda x : x.strip(' and up'))
```

```python
df[df['Current Ver']=='Varies with device'] = 0
```

```python
import re
def extract_numeric_value(text):
  text = str(text)
  numeric_value = re.findall(r'\d+\.*\d*', text)
  if numeric_value:
      return float(numeric_value[0])
  else:
      return None
# Apply the function to the 'Column' in the DataFrame
df['Current Ver'] = df['Current Ver'].apply(extract_numeric_value)
```

```python

mean_value = df['Current Ver'].mean()
df['Current Ver'].fillna(mean_value, inplace=True)
```

```python
from datetime import datetime
df['Last Updated'] = pd.to_datetime(df['Last Updated']).apply(lambda x: int(x.timestamp()))
```

```python
m_indices = df['Size'].loc[df['Size'].str.contains('M', na=False)].index.tolist()
converter = pd.DataFrame(df.loc[m_indices, 'Size'].str.replace('M', '').astype(float).apply(lambda x: x * 1024).apply(lambda x: round(x, 3)).astype(str))
df.loc[m_indices, 'Size'] = converter
```

```python
df[df['Size'] == 'Varies with device'] = 0
```

```python
for i in np.arange(1, 1001):
    size_str = str(i) + 'k'
    df.loc[df['Size'] == size_str, 'Size'] = i
```

```python
for i in np.arange(1, 1001, 0.05):
    size_str = str(round(i, 2)) + 'k'
    df.loc[df['Size'] == size_str, 'Size'] = round(i, 2)
```

```python
features = ['App', 'Reviews', 'Size', 'Installs', 'Type', 'Price', 'Content Rating', 'Genres', 'Last Updated', 'Current Ver']
features.extend(category_list)
X = df[features]
y = df['Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.18, random_state = 5)
```

```python
model = KNeighborsRegressor(n_neighbors=100)
```

```python
my_history = model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test,y_test)
'Accuracy: ' + str(np.round(accuracy*100, 2)) + '%'
```

```python
import matplotlib.pyplot as plt
```

```python
# Try different numbers of n_estimators - this will take a minute or so
n_neighbors = np.arange(1, 100, 1)
scores = []
for n in n_neighbors:
    model.set_params(n_neighbors=n)
    model.fit(X_train, y_train)
    scores.append(model.score(X_test, y_test))
plt.figure(figsize=(7, 5))
plt.title("Effect of Estimators")
plt.xlabel("Number of Neighbors K")
plt.ylabel("Score")
plt.plot(n_neighbors, scores)
```

```python
predictions = model.predict(X_test)
'MAE:', metrics.mean_absolute_error(y_test, predictions)
```

```python
'MSE:', metrics.mean_squared_error(y_test, predictions)
```
